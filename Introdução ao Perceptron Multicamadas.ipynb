# Introdução ao Perceptron Multicamadas
- Bibliotecas TensorFlow e Keras

```python
#Importação de bibliotecas
import pandas as pd
from keras.layers import Dense
from keras.models import Sequential
from keras import optimizers
import matplotlib.pyplot as plt
import numpy


#Importando dados
arquivo = "Dados1.txt"
dados = pd.read_csv(arquivo, sep="\t")
x = dados.values[:, 0]
y = dados.values[:, 1]

plt.scatter(x,y)
plt.show()
```

```python
#Estrutura da rede
model = Sequential()
layer1 = Dense(units=10, activation="tanh", input_dim=1)
layer2 = Dense(units=1, activation="linear")
model.add(layer1)
model.add(layer2)

optimizer = optimizers.SGD(lr=0.1, momentum=0.9)
model.compile(optimizer=optimizer, loss="mse")

#Parte treinamento
model.fit(x,y, epochs=1000, batch_size=32, verbose=0)


```

```python
#Valores preditos
preditos = model.predict(x)

#Plotando Resultados
plt.scatter(x, y)
plt.plot(x, preditos, color="black")
plt.show()
```
