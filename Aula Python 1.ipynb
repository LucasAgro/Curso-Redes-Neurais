```python
tempMax = [30, 29, 27, 28, 30]   #Temp máxima (°C)
tempMin = [10, 10, 10, 10, 10]   #Temp mínima (°C)
ra = [18,30,32,25,19]   #Ra (MJ m-2 d-1)


eto = []
for i in range(5):
    tempMedia = (tempMax[i] + tempMin[i])/2.0
    amplitude = tempMax[i] - tempMin[i]
    etoTemp = 0.0023*(0.408*ra[i])*(tempMedia+17.8)*(amplitude)**0.5
    eto.append(etoTemp)    

for i in range(5):
    print(eto[i])
```
